# Daily Return  - بازده روزانه


نمونه کد پایتون برای محاسبه و نمایش بازده روزانه سهام



این کد با ارسال پرامپت پایین در چت جی پی تی ایجاد شده

"
لطفا یک نمونه کد پایتون برای محاسبه ریترن یک سهم یا دارایی بنویس , از کتابخانه pandas و yfinance استفاده کن.  در پایان نمودار تغییر قیمت روزانه سهم و نمودار ریترن روزانه محاسبه شده را نشان بده
"


```python
# pip install pandas yfinance matplotlib
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# تعیین نماد سهم
symbol = 'AAPL'

# دریافت داده‌های تاریخچه سهم با استفاده از yfinance
data = yf.download(symbol, start='2022-01-01', end='2024-01-01')

# محاسبه ریترن روزانه
data['Daily_Return'] = data['Adj Close'].pct_change() * 100

# حذف ردیف‌هایی که دارای NaN هستند
data = data.dropna()

# نمودار تغییر قیمت روزانه
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(data['Adj Close'])
plt.title(f'{symbol} Daily Price')
plt.xlabel('Date')
plt.ylabel('Price')

# نمودار ریترن روزانه
plt.subplot(2, 1, 2)
plt.plot(data['Daily_Return'], color='red')
plt.title(f'{symbol} Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return (%)')

plt.tight_layout()
plt.show()
```

این نمونه هم برای یک بازه زمانی کوچکتر داشته باشیم که مقایسه نمودارها ساده تر باشه. عدد قیمت و بازده هم نمایش داده شده که بتونیم محاسبه ریاضی دستی رو هم آزمایش کنیم.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# تعیین نماد سهم (مثال: اپل)
symbol = 'AAPL'

# دریافت داده‌های تاریخچه سهم با استفاده از yfinance
data = yf.download(symbol, start='2024-01-01', end='2024-01-15')

# ایجاد ستون برای ذخیره ریترن روزانه
data['Daily_Return'] = data['Adj Close'].pct_change() * 100

# حذف ردیف‌هایی که دارای NaN هستند
data = data.dropna()

# نمودار قیمت سهم در یک بازه ۱۰ روزه
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(data['Adj Close'], marker='o')

# نوشتن قیمت‌ها روی نمودار
for i, price in enumerate(data['Adj Close']):
    plt.text(data.index[i], price, f'{price:.2f}', ha='right', va='bottom')

plt.title(f'{symbol} Stock Price (Last 10 Days)')
plt.xlabel('Date')
plt.ylabel('Price')

# نمودار ریترن روزانه
plt.subplot(2, 1, 2)
plt.plot(data['Daily_Return'], color='red', marker='o')

# نوشتن ریترن‌ها روی نمودار
for i, ret in enumerate(data['Daily_Return']):
    plt.text(data.index[i], ret, f'{ret:.2f}', ha='right', va='bottom')

plt.title(f'{symbol} Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return (%)')

plt.tight_layout()
plt.show()
```
